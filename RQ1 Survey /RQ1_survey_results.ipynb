# Quantitative Questions

```python
df['main_pl_list'] = df['(D4) What is your main programming language(s)? '].apply(lambda x: x.split(';'))
```

```python
main_pl_dict = {}
for row in df['main_pl_list']:
  for pl in row:
    if pl not in main_pl_dict:
      main_pl_dict[pl] = 0
    main_pl_dict[pl] += 1
```

```python
for col in df.columns:
  if not col.startswith('(D') and not col.startswith('(Q'):
    continue
  df[col].value_counts().plot(kind='bar', title=col)
  print( df[col].value_counts())
  plt.show()
```
